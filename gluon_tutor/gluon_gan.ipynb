## Gluon GAN
I am going take some time to write some jupyter notebooks to dive into very detail of deep learning codes. Here is the gluon generative adversarial networks.

For GAN, things actually sometimes be every complex.The core part of GAN are generator and discriminitor. Generator responsiable of generate the almost real image from **some random parameters such as from normal distribution**, it will minimize the loss of generated image and real image. At same time, discriminator will discriminate the generated image to judge it is real or fake.

```python
import matplotlib as mlp
from matplotlib import pyplot as plt
import mxnet as mx
from mxnet import gluon, autograd, nd
from mxnet.gluon import nn
import numpy as np

ctx = mx.gpu()

x = nd.random_normal(shape=(1000, 2))

# 这里乘以a就是为了让正太分布变成一个上升的趋势，你可以把这行注释掉看看效果咋样
a = nd.array([[1, 2], [-0.1, 0.5]])
b = nd.array([1, 2])
x = nd.dot(x, a) + b
y = nd.ones(shape=(1000, 1))

batch_size = 4
train_data = mx.io.NDArrayIter(x, y, batch_size, shuffle=True)

plt.scatter(x[:, 0].asnumpy(), x[:, 1].asnumpy())
plt.show()

print("the covariance matrix is: ", nd.dot(a, a.T))
```

```python
# 建造一个generator吧
net_g = nn.Sequential()
with net_g.name_scope():
    net_g.add(nn.Dense(2))

net_d = nn.Sequential()
with net_d.name_scope():
    net_d.add(nn.Dense(5, activation='tanh'))
    net_d.add(nn.Dense(3, activation='tanh'))
    net_d.add(nn.Dense(2))
    
loss = gluon.loss.SoftmaxCrossEntropyLoss()

net_g.initialize(mx.init.Normal(0.02), ctx=ctx)
net_d.initialize(mx.init.Normal(0.02), ctx=ctx)

trainer_g = gluon.Trainer(net_g.collect_params(), 'adam', {'learning_rate': 0.01})
trainer_d = gluon.Trainer(net_d.collect_params(), 'adam', {'learning_rate': 0.05})

real_label = mx.nd.ones((batch_size, ), ctx=ctx)
fake_label = mx.nd.zeros((batch_size, ), ctx=ctx)
metric = mx.metric.Accuracy()

for epoch in range(10):
    train_data.reset()
    for i, batch in enumerate(train_data):
        data = batch.data[0].as_in_context(ctx)
        noise = nd.random_normal(shape=(batch_size, 2), ctx=ctx)
        
        with autograd.record():
            real_output = net_d(data)
            err_d_real = loss(real_output, real_label)
            
            fake = net_g(noise)
            fake_output = net_d(fake.detach())
            err_d_fake = loss(fake_output, fake_label)
            err_d = err_d_real + err_d_fake
            err_d.backward()
        
        trainer_d.step(batch_size)
        metric.update([real_label, ], [real_output, ])
        metric.update([fake_label, ], [fake_output, ])

        with autograd.record():
            output = net_d(fake)
            err_g = loss(output, real_label)
            err_g.backward()
        trainer_g.step(batch_size)
    name, acc = metric.get()
    metric.reset()
    print('epoch: {}, acc: {}'.format(epoch, acc))
    
    # generate test
    noise = nd.random_normal(shape=(100, 2), ctx=ctx)
    fake = net_g(noise)
    plt.scatter(x[:, 0].asnumpy(), x[:, 1].asnumpy())
    plt.scatter(fake[:, 0].asnumpy(), fake[:, 1].asnumpy())
    plt.show()
    
```
